```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
```

```python
# 1 OPEN HW DATA: Opens output of 1a #
```

```python
# Open files of number of HW days 
df1 = pd.read_csv("Output/NumberHWdays_ERA5_Cluj-Napoca_thresh90_dur3.csv")
#df2 = pd.read_csv("[Insert file here]")
#df3 = ...

print (df1)
```

```python
#  HWMI/Number of days (annual time series) #
fig, axs = plt.subplots(1, 1, figsize=(13, 3))

axs.set_title("ERA5", fontweight='bold')

axs.plot(df1.year, df1.NumberHWDays, '-', lw=2, c='black',zorder=5, label="90pc 3day")
#axs.plot(df2.year, df2.NumberHWDays, '-', lw=2, c='red',zorder=5)

axs.set_ylabel("Number of HW days",fontsize=12)
axs.set_ylim([0,60])

axs.set_xlabel("Year")

  
axs.set_xlim([1940,2022])
axs.grid(zorder=1)
axs.legend(loc="best")

# Display the subplots
plt.show()
```
