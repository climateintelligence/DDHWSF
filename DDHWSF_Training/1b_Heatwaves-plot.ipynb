# NOTEBOOK 1B - Plot Heatwave Index
### This script opens and plots the time series of the HW data created in notbeook 1a 
### The user can open and plot multiple files, to compare cities or definitons. 

```python
# import modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
```

## STEP 1 - OPEN HW DATA. 
### Opens .csv output of 1a 

```python
# Open files of number of HW days 
df1 = pd.read_csv("Output/NumberHWdays_ERA5_Cluj-Napoca_thresh90_dur3.csv")
#df2 = pd.read_csv("[Insert file here]") # Open extra files
#df3 = ...

print (df1)
```

## STEP 2 - PLOT HW DATA IN ERA5. 

```python
fig, axs = plt.subplots(1, 1, figsize=(13, 3))

axs.set_title("ERA5", fontweight='bold')

axs.plot(df1.year, df1.NumberHWDays, '-', lw=2, c='black',zorder=5, label="90pc 3day")
#axs.plot(df2.year, df2.NumberHWDays, '-', lw=2, c='red',zorder=5) # Plot extra time series

axs.set_ylabel("Number of HW days",fontsize=12)
axs.set_ylim([0,60])

axs.set_xlabel("Year")

axs.grid(zorder=1)
axs.legend(loc="best")

# Display the subplots
plt.show()
```

## STEP 3 - PLOT HW DATA IN PAST2K.

```python
# Paleo-climate reconstruction of HWs in chosen location#

dfp = pd.read_csv("Output/NumberHWdays_past2k_Cluj-Napoca_thresh90_dur3.csv")

# DO you want to plot the whole period (0-1850) or a sample?
year_start=1600
year_end=1850

fig, axs = plt.subplots(1, 1, figsize=(13, 3))

axs.set_title("past2k", fontweight='bold', color='navy')

axs.plot(dfp.year[year_start:year_end], dfp.NumberHWDays[year_start:year_end], ls='-', lw=2, c='navy',zorder=5)
#axs.plot(df2.year, df2.NumberHWDays, '-', lw=2, c='red',zorder=5)

axs.set_ylabel("Number of HW days",fontsize=12)
axs.set_ylim([0,60])

axs.set_xlabel("Year")

axs.grid(zorder=1)

# Display the subplots
plt.show()
```
